Upscaling an image by 2 using iterpolation

```python
import numpy as np
import cv2
from skimage.measure import compare_ssim as ssim
from matplotlib import pyplot as plt
import pywt
import os
import sys
```

```python
im = cv2.imread('Task4.tif',0)
```

```python
x=0
y=0
arr=np.zeros((128,128))
for n in range(63):
    y=0
    x=x+2
    for m in range(63):
        arr[y][x]=im[m][n]
        arr[y][x+1]=im[m][n]
        y=y+1
        arr[y][x]=im[m][n]
        arr[y][x+1]=im[m][n]
        y=y+1
        
```

```python
plt.imshow(arr,'gray', interpolation = 'none')
```
